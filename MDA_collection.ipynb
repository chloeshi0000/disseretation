```python
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\'(),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)
    text = re.sub(r"[~.,%/:;?_&+*=!-@{}\[\]\$#@^]"," ", text)
    text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '', text)
    text = re.sub('[0-9]+', '', text)
    text = " ".join([w for w in text.split() if len(w) > 3])
    return text

df_M['cleaned_MDA'] = df_M['MDA_Text'].apply(clean_text)
```

```python
df_dict = pd.read_csv('Loughran-McDonald_MasterDictionary_1993-2023.csv')
positive_words = df_dict[df_dict['Positive'] != 0]['Word'].str.lower().tolist()
negative_words = df_dict[df_dict['Negative'] != 0]['Word'].str.lower().tolist()


def sentiment_analysis(text, positive_words, negative_words):
    words = text.lower().split()
    pos_count = sum(1 for word in words if word in positive_words)
    neg_count = sum(1 for word in words if word in negative_words)
    return {'Positive': pos_count, 'Negative': neg_count}


df_M['Sentiment_Score'] = df_M['cleaned_MDA'].apply(lambda text: sentiment_analysis(text, positive_words, negative_words))
print(df_M[['cleaned_MDA', 'Sentiment_Score']])
```

```python
def net_sentiment_index(text, positive_words, negative_words):
    words = text.lower().split()
    positive_count = sum(word in positive_words for word in words)
    negative_count = sum(word in negative_words for word in words)

    if positive_count + negative_count == 0:
        return 0
    return (positive_count - negative_count) / (positive_count + negative_count)
```

```python
df_M['Net_Sentiment_Index'] = df_M['cleaned_MDA'].apply(lambda text: net_sentiment_index(text, positive_words, negative_words))
```
